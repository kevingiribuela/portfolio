# Slacks histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
```

```python
ep_50mhz = pd.read_excel('50mhz.xls')
index = (ep_50mhz['Slack Range Max'] + ep_50mhz['Slack Range Min'])/2
count = ep_50mhz['Count']
fig = plt.figure(figsize = (15,10))
plt.bar(index,count, color="g")
plt.grid(True)
plt.title("Histogram at 50MHz", fontweight="bold", fontsize=20)
plt.xlabel("Slack [ns]", fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel("Endpoints", fontsize=20)
plt.yticks(fontsize=20)
plt.ylim([0,1200])
plt.show()
```

```python
ep_100mhz = pd.read_excel('100mhz.xls')
index = (ep_100mhz['Slack Range Max'] + ep_100mhz['Slack Range Min'])/2
count = ep_100mhz['Count']
fig = plt.figure(figsize = (15,10))
plt.bar(index,count, color="g")
plt.grid(True)
plt.title("Histogram at 100MHz", fontweight="bold", fontsize=20)
plt.xlabel("Slack [ns]", fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel("Endpoints", fontsize=20)
plt.yticks(fontsize=20)
plt.ylim([0, 1200])
plt.show()
```

```python
ep_200mhz = pd.read_excel('200mhz.xls')
index = (ep_200mhz['Slack Range Max'] + ep_200mhz['Slack Range Min'])/2
count = ep_200mhz['Count']
contador = 0
for i in index:
    if i<0:
        contador=contador+1

plt.figure(figsize = (15,10))
plt.bar(index[contador:],count[contador:], color="g")
plt.bar(index[0:contador], count[0:contador], color="r")
plt.grid(True)
plt.title("Histogram at 200MHz", fontweight="bold", fontsize=20)
plt.xlabel("Slack [ns]", fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel("Endpoints")
plt.yticks(fontsize=20)
plt.ylim([0, 1200])
plt.show()
```

# Utilization reports

```python
reporte = pd.read_excel('report.xlsx')
recurso = []
recurso.append(reporte["Utilization"])
recurso.append(reporte["Available"])

LUT = []
LUT.append(recurso[0][0])
LUT.append(recurso[1][0])

LUTRAM = []
LUTRAM.append(recurso[0][1])
LUTRAM.append(recurso[1][1])

FF = []
FF.append(recurso[0][2])
FF.append(recurso[1][2])

BRAM = []
BRAM.append(recurso[0][3])
BRAM.append(recurso[1][3])

DSP = []
DSP.append(recurso[0][4])
DSP.append(recurso[1][4])

IO = []
IO.append(recurso[0][5])
IO.append(recurso[1][5])

plt.figure(figsize = (10,5), dpi=400)
plt.subplot(3,2,1)
plt.pie(LUT, labels=["Used", "Free"]);
plt.title("LUT", fontweight = "bold")

plt.subplot(3,2,2)
plt.pie(LUTRAM, labels=["Used", "Free"]);
plt.title("LUTRAM", fontweight = "bold")

plt.subplot(3,2,3)
plt.pie(FF, labels=["Used", "Free"]);
plt.title("FF", fontweight = "bold")

plt.subplot(3,2,4)
plt.pie(BRAM, labels=["Used", "Free"]);
plt.title("BRAM", fontweight = "bold")

plt.subplot(3,2,5)
plt.pie(DSP, labels=["Used", "Free"]);
plt.title("DSP", fontweight = "bold")

plt.subplot(3,2,6)
plt.pie(IO, labels=["Used", "Free"]);
plt.title("IO", fontweight = "bold");

print(reporte)
```
