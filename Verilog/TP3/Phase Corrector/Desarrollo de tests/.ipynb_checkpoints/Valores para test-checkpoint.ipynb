# <u> Generaci贸n de funciones de filtrado <u>

## Creacion de funci贸n de filtro IIR

```python
def iir (x, y):
    partial_prod_B = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    partial_sum_B  = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]

    partial_prod_A = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    partial_sum_A  = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    
    # Forward
    partial_prod_B[0].value = b[0]*x[0]
    partial_prod_B[1].value = b[1]*x[1]
    partial_prod_B[2].value = b[2]*x[2]
    
    partial_sum_B[1].value  = partial_prod_B[2].fValue + partial_prod_B[1].fValue
    partial_sum_B[0].value  = partial_prod_B[0].fValue + partial_sum_B[1].fValue
    
    # Feedback
    partial_prod_A[1].value = -a[1]*y[2]
    partial_prod_A[0].value = -a[0]*y[1]
    
    partial_sum_A[1].value  = partial_prod_A[0].fValue + partial_prod_A[1].fValue
    partial_sum_A[0].value  = partial_sum_A[1].fValue  + partial_sum_B[0].fValue
    
    y[0] = partial_sum_A[0].fValue
    return y[0]
```

## Creaci贸n de filtro PI

```python
def PI_filter(x,y):
    partial_prod = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    partial_sum  = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    
    # Forward
    partial_prod[0].value = (Ki- Kp)*x[0]
    partial_prod[1].value = Kp*x[1]

    partial_sum[0].value  = partial_prod[0].fValue + partial_prod[1].fValue
    
    # Feedback
    partial_sum[1].value  = partial_sum[0].fValue  + y[1]
    
    return partial_sum[1].fValue
```

## Creaci贸n de operador partel real

```python
def real_part(x):
    x_in = [DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC), DeFixedInt(NB_INT, NB_FRAC)]
    out  = DeFixedInt(NB_INT, NB_FRAC)
    x_in[0].value = x[0]
    x_in[1].value = x[1]
    x_in[2].value = x[2]
    
    out.value = x_in[0].fValue + x_in[2].fValue
    return out.fValue
```
